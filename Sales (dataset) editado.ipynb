### Editando dataset Sales

```python
# Importando bibliotecas

import pandas as pd
import numpy as np
```

```python
# importando tabela

base_vendas = pd.read_csv("online_sales_dataset.csv")
base_vendas
```

```python
# excluindo valores negativos da coluna "Quantity"

base_vendas = base_vendas[base_vendas['Quantity'] >= 0]
#base_vendas = base_vendas[base_vendas['InvoiceDate'] <= 2024-31-10]
base_vendas
```

```python
# Supondo que 'InvoiceDate' esteja no formato datetime
base_vendas['InvoiceDate'] = pd.to_datetime(base_vendas['InvoiceDate'])
```

```python
# Filtrar as datas atÃ© 31/10/2024
base_vendas = base_vendas[base_vendas['InvoiceDate'] <= '2024-10-31']

base_vendas
```

```python
base_vendas.info()
```

```python
print(base_vendas.shape)
```

```python
print(base_vendas.isnull().sum())
```

```python
# Excluindo linhas vazias

base_vendas = base_vendas.dropna()

print(base_vendas.shape)
print(base_vendas.isnull().sum())
```

#### Analisando produtos e categorias

* AnÃ¡lise dos produtos e categorias para verificar se os produtos estÃ£o classificados corretamente.

```python
# Selecionando as categorias 'Description' e 'Category'

prod_cat = base_vendas[['Description', 'Category']]
prod_cat.head(20)
```

###  Os produtos devem ser classificados da seguinte maneira:

Furniture: Office Chair, Desk Lamp

Stationery: Blue Pen, Notebook

Accessories: Backpack, White Mug, Wall Clock, Desk Lamp

Apparel: T-shirt

Electronics: Wireless Mouse, USB Cable, Headphones

* Conforme a tabela acima, percebe-se que hÃ¡ produtos classificados erroneamente, como exemplo, 'White Mug' (canceca branca) classificado como 'Apparel' (vestuÃ¡rio) e tambÃ©m classificado como "Eletronics".


```python
# DicionÃ¡rio com os produtos categorizados corretamente

Category = {
    'T-shirt': 'Apparel',
    'Notebook': 'Stationery',
    'Wireless Mouse': 'Electronics',
    'USB Cable': 'Accessories',
    'Backpack': 'Accessories',
    'Wall Clock': 'Accessories',
    'Blue Pen': 'Stationery',
    'Office Chair': 'Furniture',
    'Desk Lamp': 'Accessories',
    'White Mug': 'Accessories',
    'Headphones': 'Electronics'
}
```

```python
# Aplicar a correÃ§Ã£o no DataFrame

base_vendas['Category'] = base_vendas['Description'].map(Category)
base_vendas
```

```python
base_vendas.groupby(["Category","Description"]).size()
base_vendas
```
